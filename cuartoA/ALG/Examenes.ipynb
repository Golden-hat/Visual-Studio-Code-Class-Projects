# Problema 1. PD:

El programa forestal de la Comunidad dispone de **T** trabajadores que se organizan en cuadrillas. Cada cuadrilla debe estar formada por un minimo de _mint_ trabajadores, y el tamaño máximo vendrá determinada por al extensión del territorio a cubrir. Deben prevenir incendios.

Hay **N** localizaciones donde son necesarios los trabajos de conservación y prevención. También se dispone de un estudio que indica el reisgo de incendio en cada una de las localizaciones dependiendo de los trabajadores que se dediquen a cada localización: $ riesgo\left(i ,t \right) $.

No hay suficientes trabajadores, se desea minimizar el riesgo total.
Hay restricciones:
- El responsable puede decidir no enviar a ningún trabajador a la localización i. $ riesgo\left(i ,0 \right) $. 
- Los trabajadores se organizan en cuadrillas con mínimo _mint_ y máximo que depende de la extensión de la localización para $ 1 \leq i \leq N $.

### Soluciones factibles

$ X = \left[ \left(x_1, x_2, ..., x_N\right) \in \mathbb{N}^{N}  \; | \; \text{mint}, \leq x_i \leq \text{maxt}, \; 1 \leq i \leq N ; \; \sum_{1 \leq i \leq N}{x_i} <= T \right] $ 

### Función objetivo a minimizar:

$ f\left(x_1, x_2, ..., x_N\right) = \sum\limits_{\left(1\leq i \leq N \right)}{\text{riesgo}\left(i,x_{i}\right)}$

### Solución óptima buscada $ \hat{x} $

$ f(\hat{x}) = \argmin\limits_{(x_1, x_2, ..., x_N) \in X}{f\left(x_1, x_2, ..., x_N\right) }$ 

### Ecuación recursiva

Partimos de la definición de la función objetivo a minimizar:

$ \argmin\limits_{(x_1, x_2, ..., x_N) \in X}{f\left(x_1, x_2, ..., x_N\right) } = \min\limits_{(x_1, x_2, ..., x_N) \in X} \sum\limits_{\left(1\leq i \leq N \right)}{\text{riesgo}\left(i,x_{i}\right)} $ 

Nos damos cuenta que, por medio de la propiedad de los sumatorios, podemos descomponer el mismo en la expresión siguiente:

$ \argmin\limits_{(x_1, x_2, ..., x_N) \in X}{f\left(x_1, x_2, ..., x_N\right) }=  \min\limits_{(x_1, x_2, ..., x_N) \in X \; | \; \text{mint} \leq x_N \leq \text{maxt} \; | \; T - X_N } \left( \left(\sum\limits_{ 1\leq i \leq N-1}{\text{riesgo}\left(i,x_{i}\right)}\right) + \text{riesgo}(N, X_N)\right)  $ 

Con esta expresión, nos damos cuenta de la recursión al ver que podemos asignar un número mínimo y máximo de trabajadores a cada parque siempre teniendo en cuenta que el total disponible se reducirá en el proceso. Deberíamos tener en cuenta entonces que para cada paso de recursión tenemos un número finito de asignaciones posibles que se ve reducido por las asignaciones que hayamos hecho en el parque anterior. Concluimos que la solución recursiva al problema viene dada por la siguiente expresión:

$
F(t,i) = \begin{cases}
0, & \text{si } i=0; \\
\min\limits_{t = 0 \; \lor \; \text{mint} \leq j \leq \text{max} \left(\text{maxt(i)}, \; t \right)}(F(i-1,t-j) + riesgo(i,t)), & \text{si } i>0;
\end{cases}
$

# Problema 3. Backtracking:

Necesitamos ofertar tartas de un rango de 5 sabores que tengan 5 capas, y que repitan como mucho M veces cada sabor, siendo $ M = 2 $. Necesitamos conseguir todas las combinaciones posibles.

Los sabores disponibles son estos:

- $ \text{sabores} = \left\{ \text{crema, moka, choco, nata, fresa} \right\} $

```python
def tartas(sabores, c, m):
  list = []; ret = []
  def backtracking(list):
    if len(list) == c:
      ret.append(list.copy())
    else:
      for i in sabores:
        list.append(i)
        if list.count(i) > m: list.pop(); continue
        backtracking(list)
        list.pop()
    return ret
  return backtracking(list)

tartas(["crema", "moka", "choco", "nata", "fresa"], 5, 2) 
```
