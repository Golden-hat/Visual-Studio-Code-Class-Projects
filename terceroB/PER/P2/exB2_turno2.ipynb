 Se propone un problema de clasificación binaria sobre un conjunto de datos que representa escenas, scene recognition dataset (id=312). Se han obtenido 299 características de diferentes imágenes que represetan dichas escenas y hay que clasificarlas como pertenecientes o no a escenas "urbanas". Por lo tanto es un problema de clasificación binaria. Hay un total de 2407 muestras que partiremos en train y validación al 70% y 30% respectivamente.

```python
from sklearn.datasets import fetch_openml

## Descarga del dataset Scene recognition
X, y = fetch_openml(data_id=312, as_frame=False, cache=True, return_X_y=True)
print(X.shape)
```

## Ejercicio

    Se debería obtener un acierto superior al 95%.

- Hasta 2 puntos:

    si **además** de todo lo anterior aplicamos un pipeline con un PCA a los datos. Se propone además de los parámetros ya mencionados arriba, variar el número de componentes de PCA así como probar diferentes valores de profundidada en el DecisionTree, no solo 1, por ejempo max_depth 1 y 2.

Primera parte:

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=23)

G = {
    "n_estimators":[50,100,150,200],
    "learning_rate":[0.01,0.1,0.2],
    "estimator__max_depth":[1]
}

GS = GridSearchCV(AdaBoostClassifier(estimator=DecisionTreeClassifier(random_state=23)), G, scoring='accuracy', refit=True, cv=5, verbose=10)

acc = GS.fit(X_train, y_train).score(X_test, y_test)
print(f'Precisión: {acc:.1%} con {GS.best_params_}')
```

Segunda parte:

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline;
from sklearn.datasets import fetch_openml

## Descarga del dataset Scene recognition
X, y = fetch_openml(data_id=312, as_frame=False, cache=True, return_X_y=True)
print(X.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=23)

pca = PCA()
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(random_state=23))
pipe = Pipeline(steps=[("pca", pca), ("ada", ada)])

G = {
    "ada__n_estimators":[50,75,100],
    "ada__learning_rate":[0.01,0.1],
    "ada__estimator__max_depth":[1,2],
    "pca__n_components": [5,10,15,20,25]
}

GS = GridSearchCV(pipe, G, scoring='accuracy', refit=True, cv=5, verbose=10)

acc = GS.fit(X_train, y_train).score(X_test, y_test)
print(f'Precisión: {acc:.1%} con {GS.best_params_}')
```
